    * Debe aceptar 3 parámetros: 
        * `bottom` - el límite inferior de los números generados
        * `ceiling` - el límite superior de los números generados
        * `count` - cuántos números generar
    * Debe devolver una matriz de números aleatorios distribuidos uniformemente

2. Llama a tu función con 2 conjuntos de parámetros a continuación:
    * bottom=10, ceiling=15, count=100

El resultado debe ser como el siguiente:



```python
def generate_random_uniform(bottom,ceiling,count):
    return uniform.rvs(bottom,ceiling-bottom, count)
   
```

```python
array1 = generate_random_uniform(10,15,100)
array1
```

```python
plt.hist(array1, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histograma de Números Aleatorios Uniformes de array1')
plt.xlabel('Valor')
plt.ylabel('Recuento')
plt.ylim(0, 120)
plt.show()
```

```python
array2 = generate_random_uniform(10,60,1000)
array2
```

```python
plt.hist(array2, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histograma de Números Aleatorios Uniformes de array2')
plt.xlabel('Valor')
plt.ylabel('Recuento')
plt.ylim(0, 120)
plt.show()
```

How are the two distributions different?

La representación de la distribución uniforme es más precisa en el segundo histograma debido al mayor número de datos y al rango más amplio.
El primer histograma muestra más variabilidad al menor número de datos y al rango más estrecho.

## Distribución Normal

1. De la misma manera en el reto de Distribución Uniforme, escriba una función que genere números aleatorios distribuidos normalmente.
1. Genera 1.000 números distribuidos normalmente con media de 10 y desviación típica de 1
1. Genera 1.000 números distribuidos normalmente con una media de 10 y una desviación típica de 50
2. Representa gráficamente las distribuciones de los datos generados.

Resultado esperado:



```python
from scipy.stats import norm 
```

```python
def generar_random_normal(mean, desv, n):
    return norm.rvs(loc=mean, scale=desv, size=n)
```

```python
normal1 = generar_random_normal(10, 1, 1000)
normal1
```

```python
plt.hist(normal1, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histograma de Números Aleatorios Uniformes de normal1')
plt.xlabel('Valor')
plt.ylabel('Recuento')
plt.show()
```

```python
normal2 = generar_random_normal(10, 50, 1000)
normal2
```

```python
plt.hist(normal2, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histograma de Números Aleatorios Uniformes de normal2')
plt.xlabel('Valor')
plt.ylabel('Recuento')
plt.show()
```

How are the two distributions different?

En la primera distribución los datos están más concentrados cerca de la media devido a que su desviación estándar, su dispersión, es más baja.
En la segunda, vemos que los datos están más dispersos dada su desviación mayor, por lo que los datos resultantes son más variables unos con otros. 
Por este motivo, también la campana del segundo gráfico es más ancha que el del primero y su media es más baja.
Podemos concluir que a medida que aumenta la desviación estándar, la probabilidad de encontrar valores alejados de la media también aumenta.

#### Función de Densidad de Probabilidad (PDF) de una Distribución Normal
La función de densidad de probabilidad (PDF) de una distribución normal describe la forma de la campana de la distribución normal, es decir, la probabilidad relativa de que una variable aleatoria tome un valor particular.

    Forma de la campana: La PDF de una distribución normal es simétrica alrededor de la media. Cuanto más grande sea la desviación estándar, más ancha y más baja será la campana.

    Área bajo la curva: El área total bajo la curva de la PDF es igual a 1, lo que significa que la suma de todas las probabilidades posibles es 1.

    Interpretación: Cuando generamos normal_pdf1, estamos calculando los valores de la PDF para los puntos x1. Si los datos en fuel_barrels_per_year se distribuyen aproximadamente de manera normal, la curva normal_pdf1 debería ajustarse bien a la distribución de los datos en un histograma.

1. Barriles de combustible/año

```python
mean1, std1 = np.mean(fuel_barrels_per_year), np.std(fuel_barrels_per_year, ddof=1)
x1 = np.linspace(min(fuel_barrels_per_year), max(fuel_barrels_per_year), 100)
normal_pdf1 = (1 / (std1 * np.sqrt(2 * np.pi))) * np.exp(-(x1 - mean1) ** 2 / (2 * std1 ** 2))
print(mean1,std1)
normal_pdf1.sum()
```

```python
plt.hist(normal_pdf1, bins=30, edgecolor='black')
plt.xlabel('normal_pdf1')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
def generate_random_exponential(n):
    return np.random.exponential(10, n)
```

```python
exp1 = generate_random_exponential(10)
```

```python
exp2 = generate_random_exponential(100)
```

```python

plt.subplot(1, 2, 1)
plt.hist(exp1, bins=100, edgecolor='black')
plt.title('Histograma - Tamaño 10')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(exp2, bins=100, edgecolor='black')
plt.title('Histograma - Tamaño 100')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

A medida que aumenta el tamaño de la muestra, la estimación de la distribución teórica mejora y las fluctuaciones aleatorias se suavizan, por lo tanto en el histograma de tamaño 100 la distirbución se acerca más a una exponencial que la del hisograma del tamaño 10. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
media = 10  # Media de la distribución exponencial (en minutos)
lambda_param = 1 / media  # Parámetro lambda

# Tiempo límite
tiempo_limite = 15  # Tiempo en minutos

# Calcular la probabilidad usando la CDF de la distribución exponencial
probabilidad_menos_quince = 1 - np.exp(-lambda_param * tiempo_limite)

print(f"La probabilidad de que un cliente pase menos de quince minutos en el banco es: {probabilidad_menos_quince:.4f}")
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
probabilidad_mas_quince = np.exp(-lambda_param * tiempo_limite)
```

```python
print(f"La probabilidad de que un cliente pase más de quince minutos en el banco es: {probabilidad_mas_quince:.4f}")
```
